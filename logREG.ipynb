## Feature Engineering/ pre procesing

```python
df["CLMSEX"].unique()
```

```python
df["CLMSEX"]=df["CLMSEX"].fillna(1)
```

```python
df["CLMINSUR"].unique()
```

```python
df["CLMINSUR"]=df["CLMINSUR"].fillna(1)
```

```python
df["SEATBELT"].unique()
```

```python
df["SEATBELT"]=df["SEATBELT"].fillna(1)
```

```python
df["CLMAGE"].unique()
```

```python
df["CLMAGE"].mean()
```

```python
df["CLMAGE"]=df["CLMAGE"].fillna(df["CLMAGE"].mean())
```

# Discribe X and Y

```python
y=df[["ATTORNEY"]]
```

```python
x=df.drop(["ATTORNEY"],axis=1)
```

# model validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,shuffle=True,random_state=10)
```

## Model building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

### for Train data

```python
model.predict(x_train)     # give x-tarin as input for find y-train
```

```python
ytrainpred=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
accuracy_score(y_train,ytrainpred)
```

### for test data

```python
ytestpred=model.predict(x_test)
```

```python
accuracy_score(y_test,ytestpred)
```

### confusion_matrix  [TP+TN/TP+TN+FP+FN]

```python
confusion_matrix(y_test,ytestpred)
```

### precision_score  [TP/TP+FP]

```python
precision_score(y_test,ytestpred)
```

### Recall  [TP/TP+FN]

```python
recall_score(y_test,ytestpred)
```
